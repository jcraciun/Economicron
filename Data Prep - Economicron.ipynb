### Adding a Continents Column <a class="anchor" id="section_1_3_1"></a>

Unfortunately, we're missing continents in this dataset. So let's merge it with a dataset involving countries from https://github.com/dbouquin/IS_608/blob/master/NanosatDB_munging/Countries-Continents.csv. 

Let's see what countries are not in that dataset, however, so that we can add them to the Excel sheet and merge properly later on. 

```python
missing_countries_check = pd.merge(combined_df, continent_list, left_on = 'COUNTRY', right_on = 'Country')
```

```python
print(len(missing_countries_check.COUNTRY.unique()))
new_country_unique = missing_countries_check.COUNTRY.unique()
```

```python
set(combined_df.COUNTRY.unique()).difference(new_country_unique)
```

These are the countries currently missing in the excel sheet. It's much faster to just update the old csv to include the new ones. 

```python
updated_continent_list = pd.read_csv('updated_countries.csv')
updated_continent_list.head(3)
```

```python
countries_continents = pd.merge(combined_df, updated_continent_list, left_on = 'COUNTRY', right_on = 'Country')
```

```python
len(countries_continents.COUNTRY.unique())
```

```python
set(combined_df.COUNTRY.unique()).difference(countries_continents.COUNTRY.unique())
```

There are now no missing countries between the datasets. 

```python
# delete the 'Country' column
countries_continents = countries_continents.drop('Country', axis = 1)
```

```python
countries_continents.head(3)
```

Let's export this once more so that we can send it to the team. 

```python
countries_continents.to_csv('countries_continents.csv', index=False)
```
